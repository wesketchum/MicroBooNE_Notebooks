```c++
for(gallery::Event ev(filenames) ; !ev.atEnd(); ev.next()) {
    //std::cout << "Processing " 
    //          << "Run " << ev.eventAuxiliary().run() << ", "
    //          << "Event " << ev.eventAuxiliary().event() << endl;
    
    auto const& mctruth_handle = ev.getValidHandle<vector<simb::MCTruth>>(gen_tag);
    auto const& truth_vec(*mctruth_handle);
    //std::cout << "\tThere are " << truth_vec.size() << " MCTruths in this event." << std::endl;

    //Loop over and print them out
    for( simb::MCTruth const& mct : truth_vec){
      
      if(mct.Origin()!=simb::Origin_t::kBeamNeutrino) continue;
      
      auto const& nu = mct.GetNeutrino();
      ccnc = nu.CCNC();
      mode = nu.Mode();
      q2 = nu.QSqr();
      nu_energy = nu.Nu().E();
      nu_pdg = nu.Nu().PdgCode();
      vtx_x = nu.Nu().Vx();
      vtx_y = nu.Nu().Vy();
      vtx_z = nu.Nu().Vz();
      
      //std::cout << nu_pdg << " " << nu_energy << " " << nu.Nu().Vx() << "," << nu.Nu().Vy() << "," << nu.Nu().Vz() << std::endl;
      
      if(mode==0) { //CC
        lep_energy = nu.Lepton().E();
        lep_p = nu.Lepton().P();
      }
      else{
        lep_energy = -999;
        lep_p = -999;
      }

      n_lep=0; n_proton=0; n_pi=0; n_pi0=0; n_gamma=0;
      vis_energy=0; qe_energy=0;
      max_proton_ke=0;
      float max_lepton_ke=0;
      
      for(size_t i_p=0; i_p<mct.NParticles(); ++i_p){
        auto const& p = mct.GetParticle(i_p);
        //std::cout << "particle " << i_p << " " << p.Mother() << " " << p.StatusCode() << " " << p.PdgCode() << " " << p.E() << " " << p.Mass() << " " << p.E()-p.Mass() << " " << p.P() << std::endl;

        if(p.StatusCode()!=1) continue;

        auto ke = p.E()-p.Mass();

        if(std::abs(p.PdgCode())==11){
            if(ke > LEP_ENERGY) ++n_lep;
            vis_energy += ke;
            if(ke>max_lepton_ke) max_lepton_ke=ke;
        }
        else if(std::abs(p.PdgCode())==13){
            if(ke > LEP_ENERGY) ++n_lep;
            vis_energy += ke;
            if(ke>max_lepton_ke) max_lepton_ke=ke;
        }
        else if(std::abs(p.PdgCode())==21){
            if(ke > GAMMA_ENERGY) ++n_gamma;
            vis_energy += ke;
        }
        else if(std::abs(p.PdgCode())==2212){
            if(ke > PROTON_ENERGY) ++n_proton;
            vis_energy += ke;
            if(ke>max_proton_ke) max_proton_ke = ke;
        }
        else if(std::abs(p.PdgCode())==211){
            if(ke > CHARGEPI_ENERGY) ++n_pi;
            vis_energy += ke;
        }
        else if(std::abs(p.PdgCode())==111){
            if(ke > PIZERO_ENERGY) ++n_pi0;
            vis_energy += p.E();
        }
      }//end loop over particles
        
      qe_energy = max_lepton_ke+max_proton_ke;
      fAnaTree->Fill();
    }//end loop over truth?

}//end loop over events
```
